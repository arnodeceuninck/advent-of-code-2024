```python
with open("input.txt") as f:
    content = f.readlines()
    content = [line.strip() for line in content]
```

```python
xmas_count = 0
```

## Horizonal

```python
# Count normal occurances
horizontal_count = 0
for line in content:
    horizontal_count += line.count('XMAS')
    horizontal_count += line.count('SAMX')

xmas_count += horizontal_count
horizontal_count
```

## Vertical

```python
transposed = []

for i in range(len(content[0])):
    vertical_row = list([content[j][i] for j in range(len(content))])
    transposed.append("".join(vertical_row))

transposed[0:5]
```

```python
vertical_count = 0
for line in transposed:
    vertical_count += line.count('XMAS')
    vertical_count += line.count('SAMX')

xmas_count += vertical_count
vertical_count
```

## Diagonal

```python
x_max = len(content[0])
y_max = len(content)
sum_max = x_max + y_max
```

```python
# all coordinates of which the some is equal should belong together
letter_per_coordinate = [] # x, y, letter (with upper left as reference)

for i in range(len(content)):
    for j in range(len(content[0])):
        letter_per_coordinate.append((i, j, content[i][j]))

letter_per_coordinate[0:5]

```

```python
diagonal_1_count = 0

for coordinate_sum in range(sum_max):
    matching_numbers = list(filter(lambda x: x[0] + x[1] == coordinate_sum, letter_per_coordinate))
    matching_numbers = sorted(matching_numbers)
    matching_letters = [x[2] for x in matching_numbers]
    matching_string = "".join(matching_letters)
    print(matching_string)
    
    diagonal_1_count += matching_string.count('XMAS')
    diagonal_1_count += matching_string.count('SAMX')

xmas_count += diagonal_1_count
diagonal_1_count
    
```

# Diagonal 2

```python
letter_per_coordinate = [] # x, y, letter (with upper left as reference)

for i in range(len(content)):
    for j in range(len(content[0])):
        letter_per_coordinate.append((i, x_max-j, content[i][j]))

letter_per_coordinate[0:5]
```

```python
diagonal_2_count = 0
for coordinate_sum in range(sum_max):
    matching_numbers = list(filter(lambda x: x[0] + x[1] == coordinate_sum, letter_per_coordinate))
    matching_numbers = sorted(matching_numbers)
    matching_letters = [x[2] for x in matching_numbers]
    matching_string = "".join(matching_letters)
    print(matching_string)
    
    if 'XMAS' in matching_string or 'SAMX' in matching_string:
        # print("Match")    
        diagonal_2_count += matching_string.count('XMAS')
        diagonal_2_count += matching_string.count('SAMX')

xmas_count += diagonal_2_count
diagonal_2_count
    
```
